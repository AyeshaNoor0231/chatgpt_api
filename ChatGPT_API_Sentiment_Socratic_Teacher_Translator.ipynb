```python
def Sentiment_analysis(text):
  messages = [
      {
      "role": "system", "content" :''' You are trained to analyse the sentiment of the text.
                                       If you are unsure of the answer you can say "Unsure" and recommend user to review manually.   ''' },
      {
          "role" : "user", "content": f''' Ananlyze the text and determine the sentiment of the text. Either it is happy, sad, overjoyed, anxious etc.
                                            Determine sentiment in one word and suggest what one should do in such emotion.
                                            Suggest a movie, exercise or food as per sentiment.
                                            Return answer in maximum 15 words: {text}  '''     }
      ]


  response = openai.ChatCompletion.create(

               model = "gpt-3.5-turbo",
               messages = messages,
               max_tokens = 150,
               n=3,#number of responses generated by system
               stop= None,
               temperature=1)

  response_text =  response.choices[0].message.content.strip().lower()

  return response_text


```

# **Model 2**

### Socratic Teacher

```python
def Socratic_teacher(text):
  messages = [
      {
      "role": "system", "content" :''' You are a Socratic tutor. Use the following principles in responding to students:


- Facilitate open and respectful dialogue among students, creating an environment where diverse viewpoints are valued and students feel comfortable sharing their ideas.
- Actively listen to students' responses, paying careful attention to their underlying thought processes and making a genuine effort to understand their perspectives.
- Guide students in their exploration of topics by encouraging them to discover answers independently, rather than providing direct answers, to enhance their reasoning and analytical skills.
- Demonstrate humility by acknowledging your own limitations and uncertainties, modeling a growth mindset and exemplifying the value of lifelong learning.   ''' },
      {
          "role" : "user", "content": f''' Help me understand the concept: {text}  '''     }
      ]


  response = openai.ChatCompletion.create(

               model = "gpt-3.5-turbo",
               messages = messages,
               max_tokens = 1024,
               n=3,
               stop= None,
               temperature=1)

  response_text =  response.choices[0].message.content.strip().lower()

  return response_text


```

```python
input = 'Help me understand types of Machine Learning, briefly'
response = Socratic_teacher(input)
print(response)
```

**example 2**

```python
input = 'Help me learn past tense in english, with examples'
response = Socratic_teacher(input)
print(response)
```

# **Translator**

```python
def translator(text, language):
  messages = [
      {
      "role": "system", "content" :'''
You will be provided with a sentence in English, and your task is to translate it into desired language.   ''' },
      {
          "role" : "user", "content": f''' translate text into desired language : {text, language}  '''     }
      ]


  response = openai.ChatCompletion.create(

               model = "gpt-3.5-turbo",
               messages = messages,
               max_tokens = 1024,
               n=3,
               stop= None,
               temperature=1)

  response_text =  response.choices[0].message.content.strip().lower()

  return response_text
```

```python
input = 'My name is Ayesha'
language = 'Urdu'
response = translator(input, language)
print(response)
```
